```python
#aprendendo a manipular uma tabela com dados nulls, preencedos eles com media ou mediana
import numpy as np

df_salarios = pd.DataFrame({
    'Nomes' : ['Luiz', 'Yuri', 'Miguel', 'Elane',  'Maria'],
    'Sal√°rio' : [10000, np.nan, 4800, np.nan, 100000]
})

df_salarios['salario_media'] = df_salarios['Sal√°rio'].fillna(df_salarios['Sal√°rio'].mean().round(2))
df_salarios
#mean() = media round(2) = arredonda
#cria uma coluna e calcula a media salarial

df_salarios['salario_mediana'] = df_salarios['Sal√°rio'].fillna(df_salarios['Sal√°rio'].median())
df_salarios
#cria uma coluna e calcula a mediana dos salrios
```

```python
df_temperaturas = pd.DataFrame({
    'Dia' : ['Segunda', 'Ter√ßa', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura' : [29, np.nan, np.nan, 27, 30]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
#ffill() preenche com o valor anterior
```

```python
df_temperaturas = pd.DataFrame({
    'Dia' : ['Segunda', 'Ter√ßa', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura' : [29, np.nan, np.nan, 27, 30]
})

df_temperaturas['preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
#ffill() preenche com o valor posterior
```

```python
import plotly.express as px
```

```python
ordem = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending = True).reset_index()

fig = px.bar(ordem,
             x = 'senioridade',
             y = 'usd',
             title = 'M√©dia Salarial Anual por Senioridade',
             labels = {'senioridade' : 'N√≠vel de Senioridade', 'usd' : 'M√©dia Salaria Anual(USD)'}
             )
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie( remoto_contagem,
    names = 'tipo_trabalho',
    values = 'quantidade',
    title = 'Propor√ß√£o dos tipos do de trabalho'
)
fig.show()
```

```python
tamanho_empresa_contagem = df_limpo['tamanho_empresa'].value_counts().reset_index()
tamanho_empresa_contagem.columns = ['Tamanho da empresa', 'Quantidade']

fig = px.pie(tamanho_empresa_contagem,
             names = 'Tamanho da empresa',
             values = 'Quantidade',
             title = 'Quantidade de empresa por tamanho')

fig.show()
```

```python
#salario por pais na vaga cientista de dados

top_10_paises = df_limpo['residencia'].value_counts().head(10).index.tolist()
# Identifica os 10 pa√≠ses com mais entradas no DataFrame e armazena seus nomes em uma lista.
df_top_10_paises = df_limpo[df_limpo['residencia'].isin(top_10_paises)]
# Um novo DataFrame contendo apenas as linhas dos 10 pa√≠ses mais frequentes.
df_Data_Scientist = df_top_10_paises[df_top_10_paises['cargo'] == 'Data Scientist']
# Filtra ainda mais o DataFrame para incluir apenas as linhas onde o 'cargo' √© 'Data Scientist'.

media_data_scientist = df_Data_Scientist.groupby('residencia')['usd'].mean().sort_values(ascending = True).reset_index()
# Calcula a m√©dia salarial em USD ('usd') para Cientistas de Dados em cada um dos 10 pa√≠ses selecionados, ordena os resultados em ordem crescente e os transforma em um DataFrame.

fig = px.bar(media_data_scientist,
             x = 'residencia',
             y = 'usd',
             title = 'Vaga Cientista de Dados: M√©dia Sal√°rial por Pa√≠s',
             labels = {'residencia' : 'Pa√≠s', 'usd' : 'Sal√°rio Anual(USD)'}
)

fig.show()
```

```python
import pycountry

# Fun√ß√£o para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar nova coluna com c√≥digo ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcular m√©dia salarial por pa√≠s (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='Sal√°rio m√©dio de Cientista de Dados por pa√≠s',
                    labels={'usd': 'Sal√°rio m√©dio (USD)', 'residencia_iso3': 'Pa√≠s'})

fig.show()
```
